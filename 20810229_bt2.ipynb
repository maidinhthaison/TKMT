1. Kiểm tra có bao nhiêu ô NA, bao nhiêu dòng có chứa NA, bao nhiêu cột có chứa NA

```python
#.3 Nạp dữ liệu
auto_names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration","model_year","origin","car_name"]
auto_mpg = pd.read_csv('auto-mpg.data', delim_whitespace=True, skiprows = 0,names=auto_names)
auto_mpg
```

```python
#4. Kiểm tra lại thông tin của bảng dữ liệu
auto_mpg.isin(['?']).any()
```

```python
# Thấy cột horsepower có dấu '?'. Lấy danh sách xe có horsepower = '?'
horsepower=auto_mpg.loc[auto_mpg['horsepower'] == '?']
horsepower
# Kết quả: có 3 loại xe mà horsepower ='?'
```

```python
# Lấy danh sách xe ford có horsepower = ?
ford_car=horsepower.loc[horsepower['car_name'].str.contains('ford')]
# Lấy horsepower của tất cả xe ford
ford_horsepower = auto_mpg.loc[(auto_mpg['car_name'].str.contains('ford', case= False)) & (auto_mpg['horsepower'] != '?')]
# Ép kiểu horsepower từ object sang float
ford_horsepower['horsepower'] = ford_horsepower['horsepower'].astype(float)
# Lấy horsepower trung bình của xe ford
avg_ford = ford_horsepower['horsepower'].mean()
# Thay thế các giá trị horsepower = ? của xe ford bằng giá trị trung bình này
indices = ford_car['horsepower'].index.values.tolist()
auto_mpg.loc[indices,'horsepower'] = np.around(avg_ford,2)
auto_mpg.head(33)
```

```python
# Lấy danh sách xe renault có horsepower = ?
renault_car=horsepower.loc[horsepower['car_name'].str.contains('renault')]
# Lấy horsepower của tất cả xe renault
renault_horsepower = auto_mpg.loc[(auto_mpg['car_name'].str.contains('renault', case= False)) & (auto_mpg['horsepower'] != '?')]
# Ép kiểu horsepower từ object sang float
renault_horsepower['horsepower'] = renault_horsepower['horsepower'].astype(float)
# Lấy horsepower trung bình của xe renault
avg_renault = renault_horsepower['horsepower'].mean()
# Thay thế các giá trị horsepower = ? của xe renault bằng giá trị trung bình này
indices = renault_car['horsepower'].index.values.tolist()
auto_mpg.loc[indices,'horsepower'] = np.around(avg_renault,2)
auto_mpg.head(331)
```

```python
# Lấy danh sách xe amc có horsepower = ?
amc_car=horsepower.loc[horsepower['car_name'].str.contains('amc')]
# Lấy horsepower của tất cả xe amc
amc_horsepower = auto_mpg.loc[(auto_mpg['car_name'].str.contains('amc', case= False)) & (auto_mpg['horsepower'] != '?')]
# Ép kiểu horsepower từ object sang float
amc_horsepower['horsepower'] = amc_horsepower['horsepower'].astype(float)
# Lấy horsepower trung bình của xe amc
avg_amc = amc_horsepower['horsepower'].mean()
# Thay thế các giá trị horsepower = ? của xe amc bằng giá trị trung bình này
indices = amc_car['horsepower'].index.values.tolist()
auto_mpg.loc[indices,'horsepower'] = np.around(avg_amc,2)
auto_mpg.head(375)
```

```python
#Kiểm tra lại
auto_mpg.isin(['?']).any()
```

```python
#5. Kiểm tra có bao nhiêu ô NA
auto_mpg.isna().sum().sum()
```

Từ 3 bảng trên, thực hiện các truy vấn dữ liệu sau (có thể viết thêm các hàm phụ nếu cần):

1.	Đưa ra danh sách các hãng cung ứng có trụ sở (CITY) ở Paris.

1.	Đưa ra tên (NAME) của các sản phẩm có trọng lượng (WEIGHT) lớn hơn 15 và có màu (COLOR) xanh (Green).

1.	Đưa ra S# của hãng có STATUS lớn nhất.

1.	Đưa ra S# của các hãng cung ứng cả hai mặt hàng P1 và P2.

1.	Đưa ra S# của các hãng cung ứng mặt hàng Nut hoặc Bolt.

1.	Đưa ra S# của các hãng cung ứng ít nhất một mặt hàng màu đỏ.

1.	Đưa ra SNAME của các hãng cung ứng ít nhất hai mặt hàng.

```python
#1 Đưa ra danh sách các hãng cung ứng có trụ sở (CITY) ở Paris.
supplier_idx = supplier.drop("S_ID", axis=1)
supplier_idx.index = supplier.S_ID

supplies_idx = supplies.drop("S_ID", axis=1)
supplies_idx.index = supplies.S_ID

result1 = supplier_idx.loc[supplier_idx['CITY']=='Paris']
result1.drop(['STATUS'], axis=1)
```

```python
#2 Đưa ra tên (NAME) của các sản phẩm có trọng lượng (WEIGHT) lớn hơn 15 và có màu (COLOR) xanh (Green).
result2= product.loc[(product['WEIGHT'] > 15) & (product['COLOR']== 'Green')]
result2.drop(['CITY'], axis=1)
```

```python
#3 Đưa ra S# của hãng có STATUS lớn nhất.
result3 = supplier[supplier['STATUS'] == supplier['STATUS'].max()]
result3.drop(['CITY'], axis=1)
```

```python
#4. Đưa ra S# của các hãng cung ứng cả hai mặt hàng P1 và P2.
result4= supplies.loc[(supplies['P_ID'] == 'P1') | (supplies['P_ID'] == 'P2' )]
result4.drop(['QTY'], axis=1)
```

```python
# 5. Đưa ra S# của các hãng cung ứng mặt hàng Nut hoặc Bolt.
product_idx = product.drop("P_ID", axis=1)
product_idx.index = product.P_ID

a = product.loc[(product['NAME'] == 'Nut') | (product['NAME'] == 'Bolt' )]
result5 = supplies.merge(a, how='inner', on='P_ID').drop(['QTY','COLOR','WEIGHT','CITY'], axis=1)
result5
```

```python

#6. Đưa ra S# của các hãng cung ứng ít nhất một mặt hàng màu đỏ.
red = product.loc[product['COLOR'] == 'Red']
result6 = supplies.merge(red, how='inner', on='P_ID').drop(['QTY','NAME','WEIGHT','CITY'], axis=1)
result6
```

```python
#7 Đưa ra S NAME của các hãng cung ứng ít nhất hai mặt hàng.
atleast2 = supplies['P_ID'].value_counts(ascending=True).loc[lambda x : x >= 2].to_frame()
atleast2_supplies= pd.merge(supplies, atleast2, how='inner', on='P_ID')
atleast2_supplies
result7 = pd.merge(supplier, atleast2_supplies, how='inner', on='S_ID')
result7.drop(['QTY', 'count','STATUS','CITY'], axis=1)
```

```python
#8. Đưa ra S# của các hãng cung ứng tất cả các mặt hàng.
result8 = pd.merge(supplier, supplies, how='inner', on='S_ID').drop(['NAME','STATUS','CITY','QTY'], axis=1)
result8.groupby(['S_ID']).apply(lambda a: a[:])
#result8.groupby(['S_ID']).apply(display)
```
