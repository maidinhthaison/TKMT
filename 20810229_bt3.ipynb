### Câu 2 (1 điểm)

1. Vẽ biểu đồ quạt cho biết tần số các nhóm của hạng hành khách (passengerClass) trong bộ dữ liệu Titanic.
1. Vẽ biểu đồ thanh cho biết tần số các nhóm của hạng hành khách.

```python
plt.style.use("seaborn-v0_8-dark-palette")
```

```python
uniquePassengerClass= titanic["passengerClass"].unique()
totalPassengerClass = titanic['passengerClass'].count()
```

```python

passengerClass= []
for i in uniquePassengerClass:
  item= titanic.loc[titanic['passengerClass'] == i]
  passengerClass.append(item)
values_pie = [x['passengerClass'].count()/totalPassengerClass * 100 for x in passengerClass]
plt.pie(values_pie, labels=uniquePassengerClass, startangle=90,autopct= '%.2f%%', counterclock=False, pctdistance=1.2, labeldistance=1.4)
plt.title("Frequency of passenger Class", y=-0.1)
plt.show()
```

```python
values_bars = [x['passengerClass'].count() for x in passengerClass]
plt.bar(uniquePassengerClass, values_bars, width=0.5, label="Passenger (persons)")

plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Frequency of passenger Class")
plt.legend()
plt.show()
```

### Câu 3 (1 điểm)

1. Vẽ biểu đồ thanh gom nhóm cho tần số sống chết (survived) chia theo hạng hành khách.
1. Vẽ biểu đồ thanh xếp chồng cho tần số sống chết chia theo hạng hành khách.

```python
uniqueSurvived= titanic["survived"].unique()
uniqueSurvived
```

```python
height = 0.4
y_pos = range(len(uniquePassengerClass))
survived_nums=[]
unsurvived_nums=[]

for value in passengerClass :
    rows = len(value.index)
    count_yes = list(value['survived']).count('yes')
    survived_nums.append(count_yes)
    unsurvived_nums.append(rows - count_yes)
y_pos1 = [y - height/2 for y in y_pos]
y_pos2 = [y + height/2 for y in y_pos]

plt.barh(y_pos1, survived_nums, height=height, label="survived")
plt.barh(y_pos2, unsurvived_nums, height=height, label="unsurvived")

plt.xlabel("Count")
plt.ylabel("Passenger Class")
plt.xticks([100*i for i in range(6)])
plt.yticks(range(len(uniquePassengerClass)), uniquePassengerClass)
plt.gca().invert_yaxis()
plt.legend()
plt.title("Survived Satisfaction per passenger class")
plt.show()
```

### Câu 6 (1 điểm)

1. Dùng histogram kiểm tra xem tuổi có ngoại lai hay không?
1. Dùng boxplot kiểm tra xem tuổi có ngoại lai hay không?
1. Tính trung bình, trung vị của tuổi và so sánh.

```python
sns.set_style("darkgrid")
```

```python
titanic['age'].isna().sum()
#Thay giá trị NA của côt age bằng giá trị trung bình
titanic['age'].fillna(value=titanic['age'].mean() , inplace=True)
titanic['age'].isna().any()
titanic['age'].describe()
```

```python
#Calculate and print outliers
q1, q3 = np.percentile(titanic['age'], [25, 75])
print(q1)
print(q3)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
outliers = [x for x in titanic['age'] if x < lower_bound  or x > upper_bound]
print("Outliers:", outliers)
```

```python
#1. Dùng histogram kiểm tra xem tuổi có ngoại lai hay không?
min = titanic['age'].min()
max = titanic['age'].max();
bins = np.round((max-min)/10)
ax = sns.histplot( data = outliers,  bins= bins.astype(int))
ax.set(xlabel='age', ylabel='outliers', title='Outliers of age')
plt.show()
#Nhận xét: Tuổi có ngoại lai khi ở độ tuổi nhỏ hơn 10 (có 34 giá trị), và từ 50t-70t (có 59 giá trị) là nhiều.
#Nhưng trong tuổi từ 10-50 thì không có giá trị ngoại lai nào
```

```python
#2. Dùng boxplot kiểm tra xem tuổi có ngoại lai hay không?
flierprops = dict(marker='o', markerfacecolor='yellow', markersize=8,  markeredgecolor='black')
sns.boxplot(x="passengerClass", y="age",
            hue="survived", palette=["m", "g"], flierprops = flierprops,
            data=titanic)
sns.despine(offset=10, trim=True)
#Nhận xét: Tuổi có ngoại lai khi ở độ tuổi nhỏ hơn 10 (có 34 giá trị), và từ 50t-70t (có 59 giá trị) là nhiều.
#Nhưng trong tuổi từ 10-50 thì không có giá trị ngoại lai nào
```

### Câu 8 (2.5 điểm)
Tập tin `WHO-COVID-19-global-data.csv` chứa dữ liệu số lượng ca nhiễm và chết do COVID-19 của các nước từ đầu năm 2020 đến gần hết tháng 9 năm 2023 (nguồn WHO: https://covid19.who.int/WHO-COVID-19-global-data.csv).

Tham khảo: https://covid19.who.int/, https://covid19.who.int/data.

1.	"Trích và mô tả" dữ liệu của Việt Nam (mã quốc gia là VN).
1.	"So sánh" dữ liệu của Việt Nam với Thái Land (mã quốc gia là TH).

```python
covid = pd.read_csv("WHO-COVID-19-global-data.csv")
covid.describe().transpose()
```

```python
#1. "Trích và mô tả" dữ liệu của Việt Nam (mã quốc gia là VN).
dfvn= covid.loc[covid['Country_code']=='VN']
dfvn.describe().transpose()
```

```python
#2. "So sánh" dữ liệu của Việt Nam với Thái Land (mã quốc gia là TH).
df= covid[covid['Country_code'].isin(['VN','TH'])]
df.describe().transpose()
```

```python
uniqueCountries= df["Country_code"].unique()
uniqueCountries
```

```python
#Người dùng có thể chỉnh số lượng cột muốn lấy để cho đồ thị đẹp hơn vì số liệu chênh lêch quá nhiều giữa các case
#Ví dụ : cols = ['New_cases','New_deaths'] hoặc cols = ['Cumulative_cases', 'Cumulative_deaths']
cols = ['New_cases', 'Cumulative_cases', 'New_deaths', 'Cumulative_deaths']

from_date = '2022-10-01'
to_date = '2022-10-31'

vn_data = df.loc[df['Country_code']=='VN']
th_data = df.loc[df['Country_code']=='TH']

mask_vn = (vn_data['Date_reported'] >= from_date) & (vn_data['Date_reported'] <= to_date)
mask_th = (th_data['Date_reported'] >= from_date) & (th_data['Date_reported'] <= to_date)

cases=[]
countries =[]
for i in  range(len(uniqueCountries)):
   cases +=  cols
   for j in range(len(cols)): 
       countries.append(uniqueCountries[i])
```

```python

vn_data_covid=[]
th_data_covid=[]
for j in range(len(cols)): 
       vn_data_covid.append(vn_data.loc[mask_vn][f'{cols[j]}'].sum())
       th_data_covid.append(th_data.loc[mask_th][f'{cols[j]}'].sum())
df1 = pd.DataFrame({'Cases': cases,
                   'Quantity': th_data_covid + vn_data_covid,
                   'Country': countries})
ax = sns.barplot(x='Cases', y='Quantity', hue='Country', data=df1,
            palette=['pink', 'orange'])

plt.title(f'Statistic from {from_date} to {to_date}')

for i in ax.containers:
    ax.bar_label(i,)
```

```python
sns.set(font_scale=1.2)

new_cases_filter = 'New_cases'
cumulative_cases_filter = 'Cumulative_cases'
new_deaths_filter = 'New_deaths'
cumulative_deathsfilter = 'Cumulative_deaths'

mask_vn = (vn_data['Date_reported'] >= from_date) & (vn_data['Date_reported'] <= to_date)
mask_th = (th_data['Date_reported'] >= from_date) & (th_data['Date_reported'] <= to_date)

vn_data_covid_new_cases = []
th_data_covid_new_cases = []

vn_data_covid_cumulative_cases = []
th_data_covid_cumulative_cases = []

vn_data_covid_new_deaths = []
th_data_covid_new_deaths = []

vn_data_covid_cumulative_deaths = []
th_data_covid_cumulative_deaths = []
for j in range(len(cols)): 
       vn_data_covid_new_cases.append(vn_data.loc[mask_vn][f'{new_cases_filter}'].sum())
       th_data_covid_new_cases.append(th_data.loc[mask_th][f'{new_cases_filter}'].sum())
       
       vn_data_covid_cumulative_cases.append(vn_data.loc[mask_vn][f'{cumulative_cases_filter}'].sum())
       th_data_covid_cumulative_cases.append(th_data.loc[mask_th][f'{cumulative_cases_filter}'].sum())
       
       vn_data_covid_new_deaths.append(vn_data.loc[mask_vn][f'{new_deaths_filter}'].sum())
       th_data_covid_new_deaths.append(th_data.loc[mask_th][f'{new_deaths_filter}'].sum())
       
       vn_data_covid_cumulative_deaths.append(vn_data.loc[mask_vn][f'{cumulative_deathsfilter}'].sum())
       th_data_covid_cumulative_deaths.append(th_data.loc[mask_th][f'{cumulative_deathsfilter}'].sum())
       

df2 = pd.DataFrame({'Countries': countries,
                   f'{new_cases_filter}': th_data_covid_new_cases + vn_data_covid_new_cases,
                   
                   f'{cumulative_cases_filter}': th_data_covid_cumulative_cases +  vn_data_covid_cumulative_cases ,
                   
                   f'{new_deaths_filter}': th_data_covid_new_deaths + vn_data_covid_new_deaths,
                   
                   f'{cumulative_deathsfilter}': th_data_covid_cumulative_deaths + vn_data_covid_cumulative_deaths
                   })
fig, axes = plt.subplots(2, 2, figsize=(18, 18))
fig.suptitle(f'Statistic from {from_date} to {to_date}', fontsize=20)

ax_newcase=sns.barplot(data=df2, x='Countries', y=f'{new_cases_filter}', ax=axes[0,0], palette=['green', 'blue'])
ax_cumulative_cases=sns.barplot(data=df2, x='Countries', y=f'{cumulative_cases_filter}', ax=axes[0,1], palette=['green', 'blue'])
ax_new_death=sns.barplot(data=df2, x='Countries', y=f'{new_deaths_filter}', ax=axes[1,0], palette=['green', 'blue'])
ax_cumulative_deaths=sns.barplot(data=df2, x='Countries', y=f'{cumulative_deathsfilter}', ax=axes[1,1], palette=['green', 'blue'])

for i in ax_newcase.containers:
    ax_newcase.bar_label(i,)

for i in ax_cumulative_cases.containers:
    ax_cumulative_cases.bar_label(i,)
    
for i in ax_new_death.containers:
    ax_new_death.bar_label(i,)
    
for i in ax_cumulative_deaths.containers:
    ax_cumulative_deaths.bar_label(i,)


```

```python

case_filter = cols[0] # có thể là cols[1] : 'Cumulative_cases', cols[2] : 'New_deaths', cols[3] : 'Cumulative_deaths'

x_vn = vn_data.loc[mask_vn]['Date_reported']
y_vn = vn_data.loc[mask_vn][f'{case_filter}']
x_th = th_data.loc[mask_th]['Date_reported']
y_th = th_data.loc[mask_th][f'{case_filter}']

plt.figure(figsize=(20,10))
plt.plot(x_vn, y_vn, label= 'Vietnam', color='red', marker='8')
plt.plot(x_th, y_th, label= 'Thailand', color='blue', marker='s')
plt.xlabel('Date')
plt.ylabel(f'{case_filter}')
plt.xticks(rotation=45)
plt.legend(title = "Countries")
plt.title(f'Numer of {case_filter} by countries from {from_date} to {to_date}' )

for x,y in zip(x_vn,y_vn):
    label = y
    plt.annotate(label, 
                 (x,y),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')
for x,y in zip(x_th,y_th):
    label = y
    plt.annotate(label, 
                 (x,y),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')
plt.show()
```
